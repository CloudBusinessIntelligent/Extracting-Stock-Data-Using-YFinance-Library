Let's Extract Stock Data Using a Python Library. Yayy!!. This is Exciting

```python
#let's install our libraries
!pip install yfinance
import yfinance as yf
import pandas as pd
```

Such a cool Library. We can extract data directly by the ticker symbol. Using the Ticker Module

Let's get some data from Apple. In under the ticker symbol AAPL

```python
#create an object that will allow us to extract data
apple = yf.Ticker('AAPL')
```

```python
#let's get some info on onject apple
apple_info = apple.info
```

```python
apple_info
```

```python
apple_info['country']
```

```python
#let's extract share price using history method and period parameter in the yfinance
apple_share_price_data = apple.history(period='max')
```

```python
apple_share_price_data
```

```python
apple_share_price_data.head()
```

```python
#let's reset index of the data and put inplace = True so the changes are in the data frame itself.
apple_share_price_data.reset_index(inplace=True)
```

```python
data = apple_share_price_data
data
```

let's plot Open price against Date



```python
data.plot(x='Date', y='Open')
```
